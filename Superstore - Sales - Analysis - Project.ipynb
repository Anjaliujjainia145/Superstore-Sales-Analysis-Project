# Final Sales Analysis

## 1 The analysis of the retail sales dataset shows that Technology is the top-performing category, generating the highest sales and profit. Products such as Copiers, Phones, and Accessories are the main profit drivers. Although Furniture generates high revenue, it has very low profit margins and even produces losses in some regions, particularly in the Central region. Office Supplies maintain stable profitability and contribute significantly to total profit.

## 2 From a regional perspective, the West and East regions are the strongest markets, contributing the majority of total sales and profit. California and New York are the most important states, while major cities like New York City and Los Angeles generate the highest profitability.

## 3 The Consumer segment is the largest customer group, contributing more than half of total sales. In shipping methods, Standard Class is the most preferred option, generating the highest sales and profit, while faster shipping methods are used less frequently.

## 4 Overall, the business performance is mainly driven by Technology products, Consumer customers, and the West and East regions, while excessive discounts and low margins in certain products reduce profitability.
